## Question 1.
How has the funding trend changed over the years? Analyze it and display the result and also plot the graph showing the trends of funding.

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce',format = '%d/%m/%Y')
```

```python
df.info()
```

```python
df=df.dropna(subset = ['Date'])
print(df.shape)
```

```python
df['Year'] = df['Date'].dt.year
df.head()
```

```python
funding_per_year = df.groupby('Year').size().reset_index(name='Number of funding')
funding_per_year
```

```python
#Plotting the graph to show the trend of funding over the year
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(funding_per_year['Year'], funding_per_year['Number of funding'])
plt.title('Trend of funding over the Year')

plt.xlabel('Year')
plt.ylabel('Number of funding')
plt.xticks(funding_per_year['Year'])
plt.show()
```

## Question 2.
Identify the top ten cities with the highest number of startups. Ensure that city names are standardized by correcting any discrepancies. For rows where multiple cities are listed separated by a "/", treat the first city as the primary location.

```python
df['CityLocation'].unique()
```

```python
df['CityLocation'] = df['CityLocation'].str.replace('New Delhi','Delhi', case=False)
df['CityLocation'] = df['CityLocation'].str.replace('bangalore','Bangalore', case=False)
df['CityLocation'] = df['CityLocation'].apply(lambda x: x.split('/')[0].strip() if isinstance(x, str) else x)
```

## Question 3.
Determine the top five locations receiving the highest percentage of funding.

```python
df['AmountInUSD'] = df['AmountInUSD'].str.replace(',','').fillna(0).astype(float)
```

```python
city_funding = df.groupby('CityLocation')['AmountInUSD'].sum().reset_index()
```

```python
top_funding = city_funding.nlargest(10,'AmountInUSD')
total_funding = city_funding['AmountInUSD'].sum()
```

```python
top_funding['Percentage'] = (top_funding['AmountInUSD']/total_funding)*100
```

```python
top_funding
```
